```python
import numpy as np
import pandas as pd

# Load the Excel file
FILEPATH = 'C:/Users/year3/Downloads/Lab Session Data.xlsx'# Use raw string to avoid unicode error

# Read all sheets from the Excel file
dataframes = pd.read_excel(FILEPATH, sheet_name=None)

# Extract the Purchase Data dataframe
purchase_data = dataframes['Purchase data']

# Display the head of the dataframe to understand its structure
#print(purchase_data.head())

# Segregate the data into matrices A and C
# Matrix A will consist of the features (Candies, Mangoes, Milk Packets)
# Matrix C will consist of the target variable (Payment)
A = purchase_data[['Candies (#)', 'Mangoes (Kg)', 'Milk Packets (#)']].values
C = purchase_data[['Payment (Rs)']].values

# Check the shapes of A and C
A_shape = A.shape
C_shape = C.shape

# Print the shapes of A and C
print('Shape of Matrix A:', A_shape)
print('Shape of Matrix C:', C_shape)

# Calculate the dimensionality of the vector space
vector_space_dimensionality = A_shape[1]  # Number of columns in A

# Calculate the number of vectors in the vector space
number_of_vectors = A_shape[0]  # Number of rows in A

# Calculate the rank of Matrix A
rank_A = np.linalg.matrix_rank(A)

# Calculate the pseudo-inverse of Matrix A
A_pseudo_inverse = np.linalg.pinv(A)

# Calculate the cost of each product available for sale using the pseudo-inverse
costs = A_pseudo_inverse @ C

# Print the results
print('Dimensionality of the vector space:', vector_space_dimensionality)
print('Number of vectors in the vector space:', number_of_vectors)
print('Rank of Matrix A:', rank_A)
print('Cost of each product available for sale:', costs)
```

```python
# Calculate the model vector X for predicting the cost of the products available with the vendor
X = A_pseudo_inverse @ C

# Print the model vector X
print('Model vector X for predicting the cost of the products:', X)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

# Assume purchase_data is already defined
purchase_data['Customer Class'] = np.where(purchase_data['Payment (Rs)'] > 200, 'RICH', 'POOR')

# Prepare the data for classification
X = purchase_data[['Candies (#)', 'Mangoes (Kg)', 'Milk Packets (#)']].values
y = purchase_data['Customer Class'].values

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize and train the classifier model
classifier = LogisticRegression()
classifier.fit(X_train, y_train)

# Predict the class for the test set
y_pred = classifier.predict(X_test)

# Print the classification report with zero_division parameter
print(classification_report(y_test, y_pred, zero_division=1))
```

```python
# Load IRCTC Stock Price data
irctc_data = dataframes['IRCTC Stock Price']

# Inspect the columns to find the correct column name for Price
print(irctc_data.columns)

# Assuming the correct column name is found, e.g., 'Price'
# Replace 'Price (Rs)' with the correct column name
price_mean = irctc_data['Price'].mean()
price_variance = irctc_data['Price'].var()

print('Mean of Price:', price_mean)
print('Variance of Price:', price_variance)

# Select price data for Wednesdays and calculate sample mean
wednesday_prices = irctc_data[irctc_data['Day'] == 'Wednesday']['Price']
wednesday_mean = wednesday_prices.mean()

print('Wednesday Mean:', wednesday_mean)
print('Population Mean:', price_mean)

# Compare means for April
april_prices = irctc_data[irctc_data['Month'] == 'Apr']['Price']
april_mean = april_prices.mean()

print('April Mean:', april_mean)
print('Population Mean:', price_mean)

# Calculate probability of making a loss over the stock
loss_prob = (irctc_data['Chg%'] < 0).mean()
print('Probability of making a loss:', loss_prob)

# Calculate probability of making a profit on Wednesday
profit_on_wed = (irctc_data[irctc_data['Day'] == 'Wednesday']['Chg%'] > 0).mean()
print('Probability of making a profit on Wednesday:', profit_on_wed)

# Conditional probability of making profit given it is Wednesday
conditional_prob = profit_on_wed / (irctc_data['Day'] == 'Wednesday').mean()
print('Conditional probability of making profit given Wednesday:', conditional_prob)

# Scatter plot of Chg% data against the day of the week
import matplotlib.pyplot as plt
import seaborn as sns

sns.scatterplot(data=irctc_data, x='Day', y='Chg%')
plt.show()

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from scipy import stats

# Load the data from the "thyroid0387_UCI" worksheet
thyroid_data = dataframes["thyroid0387_UCI"]

# Display the first few rows of the dataframe to understand its structure
print(thyroid_data.head())

# Step 2: Study Each Attribute and Identify the Datatype
# Display the data types of each column
print(thyroid_data.dtypes)

# Display summary statistics to get an idea of the data ranges and basic statistics
print(thyroid_data.describe(include='all'))

# Step 3: Identify Encoding Scheme for Categorical Attributes
# Identify categorical columns
categorical_columns = thyroid_data.select_dtypes(include=['object', 'category']).columns
print('Categorical Columns:', categorical_columns)

# Encoding scheme: Label Encoding for ordinal, One-Hot Encoding for nominal
# Example of applying Label Encoding (for ordinal variables if any)
label_encoders = {}
for col in categorical_columns:
    label_encoders[col] = LabelEncoder()
    thyroid_data[col] = label_encoders[col].fit_transform(thyroid_data[col])

# Example of applying One-Hot Encoding (for nominal variables if any)
thyroid_data = pd.get_dummies(thyroid_data, columns=categorical_columns)
print(thyroid_data.head())

# Step 4: Study Data Range for Numeric Variables
# Display summary statistics for numeric columns
numeric_columns = thyroid_data.select_dtypes(include=['int64', 'float64']).columns
print(thyroid_data[numeric_columns].describe())

# Step 5: Study the Presence of Missing Values
# Check for missing values in each column
missing_values = thyroid_data.isnull().sum()
print('Missing Values in Each Column:')
print(missing_values)

# Step 6: Study Presence of Outliers
# Plot box plots to visualize outliers
for col in numeric_columns:
    plt.figure(figsize=(10, 5))
    sns.boxplot(x=thyroid_data[col])
    plt.title(f'Box Plot for {col}')
    plt.show()

# Calculate Z-scores to identify outliers
z_scores = stats.zscore(thyroid_data[numeric_columns])
outliers = (abs(z_scores) > 3).sum(axis=0)
print('Number of Outliers in Each Column:')
print(outliers)

# Step 7: Calculate Mean and Variance (or Standard Deviation) for Numeric Variables
# Calculate mean and variance for numeric columns
means = thyroid_data[numeric_columns].mean()
variances = thyroid_data[numeric_columns].var()
std_devs = thyroid_data[numeric_columns].std()

print('Means of Numeric Columns:')
print(means)

print('Variances of Numeric Columns:')
print(variances)

print('Standard Deviations of Numeric Columns:')
print(std_devs)
```
